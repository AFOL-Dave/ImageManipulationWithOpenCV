# Description
The second project in the Super Data Science Python 3 Masterclass is simple image manipulation using the OpenCV library. In this project, a color image is loaded, converted to greyscale and several visual effects are applied.

## Import Necessary Libraries

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

## Read and Display the Image

```python
color_image = mpimg.imread("San Francisco.jpg")
plt.imshow(color_image)
```

## Get Image Shape

```python
color_image.shape
```

## Convert to Greyscale

```python
grey_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_image, cmap="gray")
```

## Save the Greyscale Image

```python
cv2.imwrite("San Francisco Grey.jpg", grey_image)
```

## Visual Effects
First stylize the image, then apply a pencil sketch effect.

```python
color_image = mpimg.imread("San Francisco.jpg")

stylized_image = cv2.stylization(color_image, sigma_s=200, sigma_r=0.8)

cv2.imshow("Me in San Francisco - Stylized", stylized_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
color_image = mpimg.imread("San Francisco.jpg")

pencil_sketch_grey, pencil_sketch_color = cv2.pencilSketch(color_image, sigma_s=60, sigma_r=0.8)

cv2.imshow("Me in San Francisco - Sketch", pencil_sketch_grey)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imshow("Me in San Francisco - Sketch", pencil_sketch_color)
cv2.waitKey()
cv2.destroyAllWindows()
```
